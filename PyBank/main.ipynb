Your task is to create a Python script that analyzes the records to calculate each of the following:

The total number of months included in the dataset.

The net total amount of Profit/Losses over the entire period.

The average of the changes in Profit/Losses over the entire period.

The greatest increase in profits (date and amount) over the entire period.

The greatest decrease in losses (date and amount) over the entire period.

```python
# Import csv and create path name

from pathlib import Path
import csv
import numpy

# Set Path
csvpath = Path('budget_data.csv')

# Set variables
num_of_month = []
net_change_list = []
total_months = 0
net_total = 0
minimum = 0
maximum = 0
prev_net = 867884

Trades = {}

# access the file 
with open(csvpath,'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    
    
    header = next(csvreader)
    
    for rows in csvreader:
        #Find the number of months in the spreadsheets
        total_months += 1 
        
        #Add numbers from spreadsheet into 
        net_gains = int(rows[1])
        dates = rows[0]
        
        Trades.update({dates: net_gains})
        
        
        #Find the Total profit from all trades
        net_total = net_total + net_gains
        net_profit = "{:,}".format(net_total)
        
        
        #Average of gain
        net_change = int(rows[1]) - prev_net
        prev_net = int(rows[1])
        net_change_list = net_change_list + [net_change]
        num_of_month = num_of_month + [rows[0]]
         
        

        
        #finding max and min numbers
        if minimum == 0:
            minimum = net_gains
        elif net_gains >= maximum:
            maximum = net_gains
        elif net_gains <= minimum:
            minimum = net_gains  
            
        min_ = "{:,}".format(minimum)
        max_ = "{:,}".format(maximum)
        
        
         
        #Getting key from value #
        def get_key(val):
            for key, value in Trades.items():
                if val == value:
                    return key

                
#Average formula (made outside the loop so that we could subtract list by -1)                
average = round(sum(net_change_list) / (len(net_change_list) - 1), 2)
avg = "{:,}".format(average)    

    
#Print Financial Analysis    
print("Financial Analysis")
print("--------------------------------------------------")
print(f"The total number of months: {total_months}")
print(f"The net profit: ${net_profit}")
print(f"The average change: ${avg}")
print(f"The greatest gain: {get_key(maximum)} (${max_})")
print(f"The greatest loss: {get_key(minimum)} (${min_})")  


   
## Output information to text file
file_to_output = Path('output.txt')
    
with open(file_to_output, "w") as txt_file:
        txt_file.write("Financial Analysis\n")
        txt_file.write("--------------------------------------------------\n")
        txt_file.write(f"The total number of months: {total_months}\n")
        txt_file.write(f"The net profit: ${net_profit}\n")
        txt_file.write(f"The average change: ${avg}\n")
        txt_file.write(f"The greatest gain: {get_key(maximum)} (${max_})\n")
        txt_file.write(f"The greatest loss: {get_key(minimum)} (${min_})\n")
    
    


                    
          
          
   
    
   
```
